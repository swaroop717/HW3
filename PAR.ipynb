```python
f = open('rosalind_par.txt')
content = f.readlines()
n = int(content[0])
array = list(map(int,content[1].rstrip().split()))
```

```python
q=array[0]
b=[]
t=[]
a=[]
```

```python
while len(array)>0:
    e= array[0]
    if e<q:
        b.append(array.pop(0))
    elif e==q:
        t.append(array.pop(0))
    else:
        a.append(array.pop(0))
```

```python
result=b+t+a
```

```python
print(*result)
```

```python
fr = open("outputpar.txt", "a")
fr.writelines([str(i)+" " for i in result])
fr.close()
```

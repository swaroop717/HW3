```python
#reading file
f = open('rosalind_bins (3).txt')
content = f.readlines()
```

```python
#reading inputs &processing
n1= int(content[0])
n2 = int(content[1])

sorted_array = list(map(int,content[2].rstrip().split()))
input_array = list(map(int,content[3].rstrip().split()))
```

```python
#to cross check input read
len(input_array), n1,n2, len(sorted_array)
```

```python
def binary_search(array, low, high, x):
    if high>=low:
        mid = (high+low)//2
        if array[mid]==x:
            return mid+1
        elif array[mid]>x:
            return binary_search(array, low, mid-1,x)
        elif array[mid]<x:
            return binary_search(array, mid+1, high, x)
    else:
        return -1
```

```python
output = [0]*(n2-1)
#looping through the elements to perform binary search
for i in range(n2-1):
    #call a function to perform binary search
    result = binary_search(sorted_array, 0, n1-1,input_array[i])
    output[i]=result
```

```python
print(*output)
```
